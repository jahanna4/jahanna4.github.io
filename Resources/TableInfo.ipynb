```python
import pandas as pd
```

```python
data = "cities.csv"
cities_df = pd.read_csv(data)
cities = cities_df.dropna()
cities
```

```python
new_df = cities[["City","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp","Wind Speed"]]
new_df
```

```python
table = new_df.to_html()
print(table)
```

```python
table_code = open("cities.html", "w", encoding="utf-8")
table_code.write(table)
table_code.close()
```
